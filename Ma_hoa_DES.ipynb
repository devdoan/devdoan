```python
def hex_to_binary(hex_string):

    # Chuyển chuỗi hexa thành một số nguyên (hệ 10)
    so_nguyen = int(hex_string, 16)

    # Chuyển số nguyên đó sang chuỗi nhị phân.
    chuoi_nhi_phan_co_prefix = bin(so_nguyen)

    # Cắt bỏ 2 ký tự đầu tiên ("0b") để lấy chuỗi nhị phân thuần túy.
    chuoi_nhi_phan = chuoi_nhi_phan_co_prefix[2:]

    # Tính toán độ dài chuẩn của chuỗi nhị phân.
    # Mỗi ký tự hexa bằng 4 ký tự nhị phân (bit).
    do_dai_can_thiet = len(hex_string) * 4

    # Thêm các số 0 vào đầu chuỗi cho đến khi đủ độ dài chuẩn.
    chuoi_nhi_phan_day_du = chuoi_nhi_phan.zfill(do_dai_can_thiet)

    return chuoi_nhi_phan_day_du

# Sử dụng đổi sang nhị phân
M_hex = "133457799BBCDFF0"
K_hex = "A0B1C2D3E4F56789"

M_binary = hex_to_binary(M_hex)
K_binary = hex_to_binary(K_hex)

print(f"Bản tin M (nhị phân): {M_binary}")
print(f"Khóa K (nhị phân): {K_binary}")
```

```python
# --- Các bảng hằng số của DES ---
IP = [
    58, 50, 42, 34, 26, 18, 10, 2,  60, 52, 44, 36, 28, 20,
    12, 4,  62, 54, 46, 38, 30, 22, 14, 6,  64, 56, 48, 40,
    32, 24, 16, 8,  57, 49, 41, 33, 25, 17, 9,  1,  59, 51,
    43, 35, 27, 19, 11, 3,  61, 53, 45, 37, 29, 21, 13, 5,
    63, 55, 47, 39, 31, 23, 15, 7
]
E = [
    32, 1, 2, 3, 4, 5,  4, 5, 6, 7, 8, 9,
    8, 9, 10, 11, 12, 13, 12, 13, 14, 15, 16, 17,
    16, 17, 18, 19, 20, 21, 20, 21, 22, 23, 24, 25,
    24, 25, 26, 27, 28, 29, 28, 29, 30, 31, 32, 1
]

S_boxes = [
    [
        [14, 4, 13, 1, 2, 15, 11, 8, 3, 10, 6, 12, 5, 9, 0, 7],
        [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
        [4, 1, 14, 8, 13, 6, 2, 11, 15, 12, 9, 7, 3, 10, 5, 0],
        [15, 12, 8, 2, 4, 9, 1, 7, 5, 11, 3, 14, 10, 0, 6, 13]
    ],
    [
        [15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0, 5, 10],
        [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
        [0, 14, 7, 11, 10, 4, 13, 1, 5, 8, 12, 6, 9, 3, 2, 15],
        [13, 8, 10, 1, 3, 15, 4, 2, 11, 6, 7, 12, 0, 5, 14, 9]
    ],
    [
        [10, 0, 9, 14, 6, 3, 15, 5, 1, 13, 12, 7, 11, 4, 2, 8],
        [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
        [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12, 5, 10, 14, 7],
        [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]
    ],
    [
        [7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
        [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
        [10, 6, 9, 0, 12, 11, 7, 13, 15, 1, 3, 14, 5, 2, 8, 4],
        [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]
    ],
    [
        [2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
        [14, 11, 2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
        [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
        [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]
    ],
    [
        [12, 1, 10, 15, 9, 2, 6, 8, 0, 13, 3, 4, 14, 7, 5, 11],
        [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
        [9, 14, 15, 5, 2, 8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
        [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]
    ],
    [
        [4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
        [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
        [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
        [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]
    ],
    [
        [13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12, 7],
        [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
        [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
        [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]
    ]
]


P = [
    16, 7, 20, 21, 29, 12, 28, 17, 1, 15, 23, 26,
    5, 18, 31, 10, 2,  8, 24, 14, 32, 27, 3, 9,
    19, 13, 30, 6, 22, 11, 4, 25
]

# Hàm XOR
def xor(bin_str1, bin_str2):
    result = ""
    for i in range(len(bin_str1)):
        if bin_str1[i] == bin_str2[i]:
            result += "0"
        else:
            result += "1"
    return result


# Áp dụng Hoán vị ban đầu (IP) cho bản tin M
M_perm = ""
for vi_tri in IP:
    M_perm += M_binary[vi_tri - 1]

# Chia bản tin đã hoán vị
L0 = M_perm[:32]
R0 = M_perm[32:]


# Mở rộng R0 từ 32 bit thành 48 bit bằng bảng E
R0_expanded = ""
for vi_tri in E:
    R0_expanded += R0[vi_tri - 1]

# XOR kết quả mở rộng với khóa con K1
XOR_result = xor(R0_expanded, K1)

# Xử lý qua các S-Box
# Chia kết quả XOR thành 8 khối 6-bit
blocks = []
for i in range(0, 48, 6):
    blocks.append(XOR_result[i : i+6])

# Áp dụng S-box cho từng khối và nối kết quả
combined_s_box_output = ""
for i in range(8):
    block_hien_tai = blocks[i]
    s_box_hien_tai = S_boxes[i]

    # Xác định hàng (dùng bit đầu và bit cuối)
    row = int(block_hien_tai[0] + block_hien_tai[5], 2)
    # Xác định cột (dùng 4 bit ở giữa)
    col = int(block_hien_tai[1:5], 2)

    # Tra giá trị trong S-box
    value = s_box_hien_tai[row][col]

    # Chuyển giá trị thành chuỗi 4-bit và cộng vào kết quả
    combined_s_box_output += bin(value)[2:].zfill(4)

# Áp dụng hoán vị P cho kết quả từ S-box
P_output = ""
for vi_tri in P:
    P_output += combined_s_box_output[vi_tri - 1]

# Kết thúc hàm f, P_output chính là kết quả của f(R0, K1)

# Tính L1 và R1 cho Vòng 1
L1 = R0
R1 = xor(L0, P_output)


print(f"Bản tin M sau hoán vị ban đầu (IP): {M_perm}")
print(f"L0: {L0}")
print(f"R0: {R0}")
print(f"R0 sau mở rộng (E): {R0_expanded}")
print(f"Kết quả XOR R0_expanded với K1: {XOR_result}")
print(f"Đầu vào các S-box (8 khối 6-bit): {blocks}")
print(f"Kết quả sau khi kết hợp đầu ra S-box: {combined_s_box_output}")
print(f"Kết quả sau hoán vị P (tức là f(R0, K1)): {P_output}")
print(f"L1: {L1}")
print(f"R1: {R1}")
```
