Đây là giải thích ý nghĩa của từng cột trong bộ dữ liệu:

- customerID: ID duy nhất cho mỗi khách hàng.
- gender: Giới tính của khách hàng (Nam hoặc Nữ).
- SeniorCitizen: Cho biết khách hàng có phải là công dân cao tuổi hay không (1 cho Có, 0 cho Không).
- Partner: Cho biết khách hàng có bạn đời hay không (Yes hoặc No).
- Dependents: Cho biết khách hàng có người phụ thuộc hay không (Yes hoặc No).
- tenure: Số tháng khách hàng ở lại với công ty.
- PhoneService: Cho biết khách hàng có dịch vụ điện thoại hay không (Yes hoặc No).
- MultipleLines: Cho biết khách hàng có nhiều đường dây điện thoại hay không (Yes, No hoặc No phone service).
- InternetService: Nhà cung cấp dịch vụ internet của khách hàng (DSL, Fiber optic, hoặc No).
- OnlineSecurity: Cho biết khách hàng có dịch vụ bảo mật trực tuyến hay không (Yes, No, hoặc No internet service).
- OnlineBackup: Cho biết khách hàng có dịch vụ sao lưu trực tuyến hay không (Yes, No, hoặc No internet service).
- DeviceProtection: Cho biết khách hàng có dịch vụ bảo vệ thiết bị hay không (Yes, No, hoặc No internet service).
- TechSupport: Cho biết khách hàng có dịch vụ hỗ trợ kỹ thuật hay không (Yes, No, hoặc No internet service).
StreamingTV: Cho biết khách hàng có dịch vụ truyền hình trực tuyến hay không (Yes, No, hoặc No internet service).
- StreamingMovies: Cho biết khách hàng có dịch vụ xem phim trực tuyến hay không (Yes, No, hoặc No internet service).
- Contract: Loại hợp đồng của khách hàng (Month-to-month, One year, Two year).
- PaperlessBilling: Cho biết khách hàng có sử dụng hình thức thanh toán không giấy tờ hay không (Yes hoặc No).
- PaymentMethod: Phương thức thanh toán của khách hàng (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic)).
- MonthlyCharges: Số tiền khách hàng phải trả hàng tháng.
- TotalCharges: Tổng số tiền khách hàng đã trả.
- Churn: Cho biết khách hàng có rời bỏ dịch vụ hay không (Yes hoặc No). Đây là biến mục tiêu.

## Kiểm ta dữ liệu thiếu qua đồ thị

```python
import missingno as msno
msno.matrix(df)
```

## Kiểm tra xem có ô dữ liệu nào bị thiếu không vì có thể nó có dạng string trắng

```python
for col in df.select_dtypes(include='object').columns:
    if (df[col].astype(str).str.strip() == '').any():
        print(f"Column '{col}' contains blank spaces or empty strings.")
        # Optionally display rows with blank spaces in this column
        # display(df[df[col].astype(str).str.strip() == ''])
```

```python
blank_spaces_count = df[df['TotalCharges'].astype(str).str.strip() == ''].shape[0]
print(f"Number of blank spaces or empty strings in 'TotalCharges' column: {blank_spaces_count}")
```

```python
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

## Gán NaN cho các ô trắng

```python
import numpy as np
df[np.isnan(df['TotalCharges'])]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 6))

# Plot all data points with some transparency
sns.scatterplot(data=df.dropna(subset=['TotalCharges']), x='tenure', y='TotalCharges', alpha=0.7, label='Valid TotalCharges')

# Filter for rows where TotalCharges is NaN
nan_totalcharges_df = df[np.isnan(df['TotalCharges'])]

# Plot the points with NaN TotalCharges (they will have tenure = 0)
sns.scatterplot(data=nan_totalcharges_df, x='tenure', y='MonthlyCharges', color='red', s=50, label='TotalCharges is NaN (using MonthlyCharges for y-axis)')


plt.title('Relationship between Tenure and TotalCharges (NaN values highlighted)')
plt.xlabel('Tenure (months)')
plt.ylabel('TotalCharges')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='Churn', hue='Churn', palette=['skyblue', 'salmon'], legend=False) # Use hue to apply palette correctly and suppress warning
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Count')

# Add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
